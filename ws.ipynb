```python
import requests
from bs4 import BeautifulSoup

url = "https://webscraper.io/"
r = requests.get(url)

# Check if request was successful
if r.status_code == 200:
    soup = BeautifulSoup(r.text, 'html.parser')
    print(soup.prettify())  # Prettify makes the HTML more readable
else:
    print(f"Error: Unable to fetch page, Status Code: {r.status_code}")
```

```python
# to extract all the div
print(soup.div)
```

```python
# to get atttributess
tag=soup.header
print(tag.attrs)
```

```python
# to get class
atb=(tag.attrs)
print(atb['class'])
```

```python
# to get navigational string
tag=soup.header.div.a.button.span
print(tag.string)
```
